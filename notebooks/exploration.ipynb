```python
# Data Exploration and Visualization


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the raw data
data_path = 'Amazon-Products - online (1).csv'
data = pd.read_csv(data_path)

# Display the first few rows and column names
print(data.head())
print("Columns in the dataset:", data.columns.tolist())

# Clean and preprocess the data
# Convert 'actual_price' and 'discount_price' to numeric
data['actual_price'] = data['actual_price'].replace({'₹': '', ',': ''}, regex=True).astype(float)
data['discount_price'] = data['discount_price'].replace({'₹': '', ',': ''}, regex=True).astype(float)

# Convert 'date' to datetime format
data['date'] = pd.to_datetime(data['date'], errors='coerce')

# Check if 'date' column exists and add year, month, day columns
if 'date' in data.columns:
    data['Year'] = data['date'].dt.year
    data['Month'] = data['date'].dt.month
    data['Day'] = data['date'].dt.day

    # Plot the number of records per year
    plt.figure(figsize=(12, 6))
    sns.countplot(x='Year', data=data)
    plt.title('Number of Records Per Year')
    plt.xlabel('Year')
    plt.ylabel('Count')
    plt.show()

    # Plot the number of records per month
    plt.figure(figsize=(12, 6))
    sns.countplot(x='Month', data=data)
    plt.title('Number of Records Per Month')
    plt.xlabel('Month')
    plt.ylabel('Count')
    plt.show()
else:
    print("Error: 'date' column not found.")

# Display basic statistics of numerical columns
print(data.describe())

# Display correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```
